# identify healpix and target id of posteriors that didn't run correctly

```python
import os
import h5py 
import glob
import numpy as np
```

```python
users = ['msiudek']#'chahah', 'shuang89', 'csaulder', 'msiudek']

bad_hpixs, bad_targids = [], [] 
for user in users:
    print(user)
    dat_dir = '/global/cscratch1/sd/%s/provabgs/svda/healpix' % user
    hpixs = [int(os.path.basename(subdir)) for subdir in glob.glob(os.path.join(dat_dir, '*'))]
    for hpix in hpixs:
        fposts = glob.glob(os.path.join(dat_dir, str(hpix), 'provabgs.*hdf5'))

        for fpost in fposts: 
            # check if files open
            try: 
                f = h5py.File(fpost, 'r')
                f.close()
            except: 
                targid = int(os.path.basename(fpost).split('.')[1])
                print(hpix, targid)
                bad_hpixs.append(hpix)
                bad_targids.append(targid)
```
